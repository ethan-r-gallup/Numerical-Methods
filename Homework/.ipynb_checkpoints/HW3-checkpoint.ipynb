```python
def thomas(ml,md,mu,mb):
    l = ml.copy()
    d = md.copy()
    u = mu.copy()
    b = mb.copy()
    n = len(d)
    x = np.zeros(n)
    for i in range(1,n):
        d[i] = d[i] - u[i-1] * l[i-1]/d[i-1]
        b[i] = b[i] - b[i-1] * l[i-1]/d[i-1]
    x[n-1] = b[n-1]/d[n-1]
    for i in range(n-2,-1,-1):
        x[i] = (b[i]-u[i]*x[i+1])/d[i]
    return x
```

## Problem Setup

```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
np.set_printoptions(precision=8)
```

```python
def source(x, σ, x0, α):
    return α*np.exp(-(x-x0)**2/(σ**2))
```

```python
L = 1.0
TL = 300
TR = 350
k = 1e-5
a = 2*L/3
α = 0.01
σ = 0.1
```

## Problem 1


```python
n = 100
dx = L /(n-1)
x = np.linspace(0,L,n)

main = np.ones(n) * -2
main[0], main[-1] = 1.0, 1.0

upper = np.ones(n-1)
lower = np.ones(n-1)
upper[0], lower[-1] = 0.0, 0.0
A = np.diag(lower,-1) + np.diag(main,0) + np.diag(upper,1)
```

### a)

```python
# build the RHS
b = - dx*dx*source(x, σ, a, α)/k
b[0] = TL
b[-1] = TR

# solve using build in solver.
# method 1: use the solve routine
T = np.linalg.solve(A,b)
plt.plot(x,T)
plt.title("np solve")
plt.xlabel("x (m)")
plt.ylabel("T (K)")
plt.grid()
```

The solution makes sense because the plot shows the correct temperatures on the right and left sides and shows a smooth, continuous curve between them. It also has a peak between the two ends which is what we would expect since the rod is being heated "somewhere along its length".

### b)

```python
Ainv = np.linalg.inv(A)
T = Ainv@b
plt.plot(x,T)
plt.title("Inverse")
plt.xlabel("x (m)")
plt.ylabel("T (K)")
plt.grid()
```

The solution makes sense because the plot shows the correct temperatures on the right and left sides and shows a smooth, continuous curve between them. It also has a peak between the two ends which is what we would expect since the rod is being heated "somewhere along its length". It is also identical to the previous solution.

### c)

```python
T_thomas = thomas(lower, main, upper, b)
plt.plot(x,T_thomas)
plt.title("Thomas Algorithm")
plt.xlabel("x (m)")
plt.ylabel("T (K)")
plt.grid()
```

The solution makes sense because the plot shows the correct temperatures on the right and left sides and shows a smooth, continuous curve between them. It also has a peak between the two ends which is what we would expect since the rod is being heated "somewhere along its length". It is also nearly identical to the two previous solutions.

### d)

The difference should be practically zero for the Thomas algorithm. It is a direct method for solving linear systems so it will not have any iteration error.
