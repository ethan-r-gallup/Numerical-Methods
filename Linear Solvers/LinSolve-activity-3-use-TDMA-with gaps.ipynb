Use the thomas algorithm supplied below to solve the system of equations:
\begin{equation}
\begin{bmatrix}
2 &-3  & 0 & 0\\ 
1 & 2 & -1 & 0\\ 
0 & 4 & -1 & 1\\ 
0 & 0 &  2& -1
\end{bmatrix}
\begin{pmatrix}
x_1\\ 
x_2\\ 
x_3\\ 
x_4
\end{pmatrix}
=
\begin{pmatrix}
-4\\ 
2\\ 
9\\ 
2
\end{pmatrix}
\end{equation}

```python
import numpy as np
# l_ = lower diagonal vector
# d_ = main diagonal vector
# u_ = upper diagonal vector
# b_ = right-hand-side vector
def thomas(l_ ,d_ ,u_ ,b_):
    lower = l_.copy()
    main = d_.copy()
    upper = u_.copy()

    b = b_.copy()
    n = len(main)
    x = np.zeros(n) # define solution vector
    for i in range(1,n):
        main[i] = main[i] - upper[i-1]*lower[i-1]/main[i-1]
        b[i] = b[i] - b[i-1]*lower[i-1]/main[i-1]

    x[n-1] = b[n-1]/main[n-1]
    for i in range(n-2,-1,-1):
        x[i] = (b[i]-upper[i]*x[i+1])/main[i]
    return x
```

```python
ld = [1, 4, 2]
md = [2, 2, -1, -1]
ud = [-3, -1, 1]
rhs  = [-4, 2, 9, 2]
sol = thomas(ld, md, ud, rhs)
print(sol)
```

## Verification
To verify our answer, we can use the built-in `solve` function from `numpy.linalg`

```python
from numpy.linalg import solve
# build the full coefficient matrix
A = [[2, -3, 0, 0], [1, 2, -1, 0], [0, 4, -1, 1], [0, 0, 2, -1]]
sol2 = solve(A,rhs)
print(sol2)
```
