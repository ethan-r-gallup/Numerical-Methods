```python
def newton(func, x0, tol=1e-50):
    niter = 0
    err = 1e5
    while err > tol and niter < 20:
        h = 1e-5*x0 + 1e-12
        df = (func(x0+h) - func(x0))/h
        x1 = x0 - func(x0)/df
        x0 = x1
        err = abs(func(x1))
        niter +=1
    return x1, err, niter

def f_res(f):
    global ϵ, D, Re
    return 1/sqrt(f) + 2*log10(ϵ/(3.7*D) + 2.51/(Re*sqrt(f)))
```
